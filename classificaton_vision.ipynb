```python
import tensorflow as tf
import numpy as np
from bokeh.plotting import figure, output_notebook, show
from bokeh.layouts import gridplot
output_notebook()
print(tf.__version__)
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat','Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
p = figure(plot_width=300,plot_height=300)
p.image(image=[train_images[0][::-1]],x=0,y=0,dw=10,dh=10, palette='Plasma256', level="image")
show(p)
```

```python
train_images = train_images/255.0
test_images = test_images/255.0
```

```python
offset = 10
grid_list = []
for i in range(offset,offset+16):
    p = figure(plot_width=150,plot_height=150,title=class_names[train_labels[i]],title_location='below')
    p.image(image=[train_images[i][::-1]],x=0,y=0,dw=10,dh=10,palette='Greys256',level='image')
    grid_list.append(p)
show(gridplot(grid_list, ncols=4))
```

```python
layers = [tf.keras.layers.Flatten(input_shape=train_images[0].shape)]
layers.append(tf.keras.layers.Dense(128, activation='relu'))
layers.append(tf.keras.layers.Dense(10))
layers.append(tf.keras.layers.Softmax())
model = tf.keras.Sequential(layers)
```

```python
model.compile(optimizer='adam',
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
             metrics=['accuracy'])
```

```python
model.fit(train_images,train_labels, epochs=10)
```

```python
test_loss, test_acc = model.evaluate(test_images,test_labels, verbose=2)
print(test_loss, test_acc)
```
